```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
with open('VeriOnce.csv', 'w') as file:
            fieldnames = ['ilan', 'fiyat','oda sayısı','metre kare','yas','kat']
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()   
    
            toplam = 0
            for sayfaNo in range(1,35):
                #veri çekeceğimiz web sitemizin adresi
                url="https://www.hurriyetemlak.com/izmir-satilik/daire?p7=101202-101201-101236-101215-101216-101217-101218-101219-101220-101221-101223-101225-101226-101224-101222&page="+str(sayfaNo)
                sayfa = requests.get(url)
                #requests ile url ye bağlanıyoruz
                icerik=sayfa.content
                #Yazdığımız urlden gelen html şeklindeki veriyi sayfaya kaydettik
                soup=BeautifulSoup(icerik,'html.parser')
                #sayfa kaynağından ilanlar alındı.
                ilanlar = soup.find_all("div",attrs={"class":"listing-item"})
                for ilan in ilanlar:
                    #tüm ilanlardan ilan ismi,fiyat,oda sayısı, metre kare,bina yaşı,katbilgileri alındı.
                    ilan_isim= ilan.find("div",attrs={"class":"list-view-header"}).text.strip()
                    print(ilan_isim)
                    toplam+=1
                   
                    
                    fiyat = ilan.find("div",attrs={"class":"list-view-price"}).text.strip()
                    print(fiyat)
                   
                    
                    oda_sayısı = ilan.find("span",attrs={"class":"celly houseRoomCount"}).text.strip()
                    print(oda_sayısı)
                   
                    
                    metre_kare = ilan.find("span",attrs={"class":"celly squareMeter list-view-size"}).text.strip()
                    print(metre_kare)
                        
                   
                    yas = ilan.find("span",attrs={"class":"celly buildingAge"}).text.strip()
                    print(yas)
                   

                    try:
                        kat = ilan.find("span",attrs={"class":"celly floortype"}).text.strip()
                        print(kat)
                    except Exception:
                        kat=""
                        print(kat)
                    writer.writerow({"ilan": ilan_isim, "fiyat":fiyat,"oda sayısı":oda_sayısı,"metre kare":metre_kare,"yas":yas,"kat":kat})    
print(toplam)     
```
